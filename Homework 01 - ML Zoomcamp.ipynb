```python
##Calculate mean value of df1
df2["MSRP"].mean()
```

```python
##Filter Dataset having year greater or equal to 2015 to a new Dataset named df3
df_mask = df1["Year"] >= 2015
positions = np.flatnonzero(df_mask)
df3=df1.iloc[positions]
df3
```

```python
##Verify years of the new Dataset df3
uniquedf = df3['Year'].unique()
print(sorted(uniquedf))
```

```python
##Number of rows that has Engine HP as Null Value
sum(pd.isnull(df3['Engine HP']))
```

```python
##Calculate Mean Value of Engine HP with Null Values
mean_hp_before = df3["Engine HP"].mean()
mean_hp_before
```

```python
##Replace Null Values
df3["Engine HP"].fillna(value=0, inplace=True)
```

```python
##Calculate new mean value
mean_hp_after = df3["Engine HP"].mean()
mean_hp_after
```

```python
##Compare both mean values
print(round(mean_hp_before))
print(round(mean_hp_after))
```
